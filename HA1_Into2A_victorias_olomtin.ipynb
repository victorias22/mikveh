למה בחרת את הפרויקט:
הפרויקט נבחר מתוך עניין אישי בתחום של למידת מכונה ותחום חיזוי מחירים. בנוסף, מדובר בבעיה בעלת השפעה על תחום הנדל"ן, שנראה שיכול לתרום להבנה טובה יותר של מגמות השוק ולתכנונים כלכליים. כמו כן, הפרויקט מאפשר לי להתנסות בהשוואת אלגוריתמים שונים ולהבין את היתרונות והחסרונות של כל אלגוריתם בנוגע לבעיה מסוימת.

היישום המתוכנן:
ייבוא נתונים: הנתונים ייבאו מ-Kaggle ויוטענו לסביבת Python.
עיבוד נתונים:
ניקוי נתונים: טיפול בערכים חסרים (למשל, החלפתם בערך ממוצע או שימוש בהסרת שורות) ובנתונים לא תקינים.
ניתוח משתנים חסרים: יש לבדוק אם יש משתנים חסרים או בעיות נתונים אחרות ולבצע עיבוד מתאים (למשל, חישוב ממוצעים או חיזוי).
יצירת משתנים חדשים: במידת הצורך, ניצור משתנים חדשים שיכולים לעזור לאלגוריתמים (כגון יצירת משתנים נלווים כמו יחס מחיר-גודל או קטגוריות שיכולות להקל על המודל להבין את הנתונים).
המודלים:
נבחרו אלגוריתמים שונים ללמידת מכונה, כולל Naive Bayes, SVM ו-לוגיסטיק רגרסיה, לצורך חיזוי מחירי דיור.
כל אחד מהמודלים ייושם על הנתונים, ובסופו של תהליך נבצע השוואה בין ביצועי המודלים בעזרת מדדים כמו דיוק (accuracy), אורך זמן הריצה, ועוד.
הערכה וניתוח תוצאות: לאחר ביצוע המודלים, נבחן את תוצאות כל אלגוריתם (לדוגמה, דיוק, רגישות, ספציפיות) וננתח אילו אלגוריתמים הם היעילים ביותר לבעיה זו.
מסקנות: בסופו של הפרויקט נביא את מסקנותינו לגבי האלגוריתם האופטימלי עבור בעיית חיזוי מחירי דיור, תוך שמירה על השקיפות בתהליך קבלת ההחלטות.
